```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.graphics.mosaicplot import mosaic
import seaborn as sns
from datetime import datetime

import utils

%matplotlib inline
```

```python
train_data = pd.read_csv("train.csv", index_col='AnimalID')
test_data = pd.read_csv("test.csv", index_col='ID')
all_data = pd.concat([train_data, test_data])
print(train_data.shape)
print(test_data.shape)
```

```python
color = train_data[['OutcomeType', 'Color']]
grouped = color[color['OutcomeType'] == 'Adoption'].groupby(['Color', 'OutcomeType'])
plt.rcParams['font.size'] = 12.0
plt.rcParams['figure.figsize'] = 20, 100
grouped.size().plot.barh()
```

```python
train_data = utils.prepare_age(train_data)
```

```python
def bar(t, o):
    age = train_data[['OutcomeType', 'AgeuponOutcome', 'AnimalType']]
    cat = age[age['AnimalType'] == t]
    grouped = cat[cat['OutcomeType'] == o].groupby(['OutcomeType', 'AgeuponOutcome'])
    plt.rcParams['font.size'] = 12.0
    plt.rcParams['figure.figsize'] = 20, 10
    return grouped.size().plot.bar()
```

```python
bar('Cat', 'Adoption')
```

```python
bar('Dog', 'Adoption')
```
